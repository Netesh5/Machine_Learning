```python
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
p=sb.get_dataset_names()
print(p)
```

```python
data=sb.load_dataset("planets")
print(data)
```

```python
sb.lineplot(x="mass",y="distance",data=data)
plt.show()
```

```python
sb.barplot(x="mass",y="distance",data=data)
plt.show()
```

```python
sb.countplot(x="distance",data=data)
plt.title("Planet")
```
