```python
import pandas as pd
import seaborn as sns
```

```python
data=sns.load_dataset("penguins")
data.head(10)
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True
            )
data
```

```python
data.isnull().sum()
```

**Feature Engineering / Converting the categorical data into numeric**

```python
pd.get_dummies(data["sex"]).head()
```

```python
Gender=pd.get_dummies(data["sex"],drop_first=True)
Gender
```

```python
data.island.unique
```

```python
pd.get_dummies(data["island"])
```

```python
Island=pd.get_dummies(data["island"],drop_first=True)
Island
```

```python
data.drop(["island","sex"],inplace=True,axis=1)
data
```

```python
data=pd.concat([data,Gender,Island],axis=1)
data
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
x=data.drop("species",axis=1)
x
```

```python
le=LabelEncoder()
y=le.fit_transform(data["species"])
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier(n_estimators=10,criterion="entropy",random_state=0)
model.fit(x_train,y_train)
```

```python
predict=model.predict(x_test)
predict
```

```python
score=accuracy_score(y_test,predict)
score
```

```python
confusion_matrix(y_test,predict)
```

```python
classification_report(y_test,predict)
```

**Criterion=GINI**

```python
model=RandomForestClassifier(n_estimators=7,criterion="gini",random_state=0)
model.fit(x_train,y_train)
```

```python
predict=model.predict(x_test)
predict
```

```python
score=accuracy_score(y_test,predict)
score
```
