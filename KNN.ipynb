```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
data=load_breast_cancer()
data.keys()
```

```python
print(data["DESCR"])
```

**Creating the DataFrame for the data**

```python
dataset=pd.DataFrame(data["data"],columns=data["feature_names"])
dataset.head(10)
```

```python
dataset.isnull().sum()
```

```python
dataset.info()
```

```python
dataset.shape
```

```python
dataset.describe()
```

```python
target=data["target"]
target
```

```python
target=pd.DataFrame(data["target"],columns=["cancer"])
target
```

**Scaling the feature data**

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()

scaled_data=sc.fit_transform(dataset)
scaled_data
scaled_df=pd.DataFrame(scaled_data,columns=dataset.columns)
scaled_df
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(scaled_df,np.ravel(target),test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn import neighbors
model=KNeighborsClassifier(n_neighbors=5)
model.fit(x_train,y_train)
```

```python
predict=model.predict(x_test)
predict
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
score=accuracy_score(y_test,predict)
score
```

```python
confusion_matrix(y_test,predict)
```

```python
print(classification_report(y_test,predict))
```

**Selecting the best K vlaue**

```python
for i in range(1,50):
  model=KNeighborsClassifier(n_neighbors=i)
  model.fit(x_train,y_train)
  predict=model.predict(x_test)
  score=accuracy_score(y_test,predict)
  print(i,": ", score)
```
